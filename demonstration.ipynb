### **For us
#### this is a comperason between sklearn pca andour fun of pca  

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Get numeric data
df_numeric = df.drop(columns='ballot_code').select_dtypes(include=['int64', 'float64'])

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_numeric)

pca = PCA(n_components=2)
sklearn_pca_data = pca.fit_transform(scaled_data)

# Apply your PCA
your_pca_data = files.dimensionality_reduction(df_numeric, 2, [])

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot scikit-learn PCA
ax1.scatter(sklearn_pca_data[:, 0], sklearn_pca_data[:, 1], alpha=0.5)
ax1.set_xlabel('First Principal Component')
ax1.set_ylabel('Second Principal Component')
ax1.set_title('Scikit-learn PCA')

# Plot your PCA
ax2.scatter(your_pca_data['PC1'], your_pca_data['PC2'], alpha=0.5)
ax2.set_xlabel('First Principal Component')
ax2.set_ylabel('Second Principal Component')
ax2.set_title('Your PCA Implementation')

plt.tight_layout()
plt.show()

# Print explained variance ratio from scikit-learn PCA
print("Explained variance ratio:", pca.explained_variance_ratio_)
```

## Visualize the Reduced Data

```python
agg_data = files.group_and_aggregate_data(df.drop(columns='ballot_code'), 'city_name', 'sum')

sparse_remove_agg = files.remove_sparse_columns(agg_data, 1000)

df_reduced_agg = files.dimensionality_reduction(sparse_remove_agg, 2, [])

# For visualization:
fig, ax = plt.subplots()
ax.scatter(df_reduced_agg['PC1'], df_reduced_agg['PC2'], alpha=0.5)
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()

print('We estimate that the number of clusters Visually is : 1')
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(sparse_remove_agg)

pca = PCA(n_components=2)
sklearn_pca_data = pca.fit_transform(scaled_data)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot scikit-learn PCA
ax1.scatter(sklearn_pca_data[:, 0], sklearn_pca_data[:, 1], alpha=0.5)
ax1.set_xlabel('First Principal Component')
ax1.set_ylabel('Second Principal Component')
ax1.set_title('Scikit-learn PCA')

# Plot your PCA
ax2.scatter(df_reduced_agg['PC1'], df_reduced_agg['PC2'], alpha=0.5)
ax2.set_xlabel('First Principal Component')
ax2.set_ylabel('Second Principal Component')
ax2.set_title('Your PCA Implementation')

plt.tight_layout()
plt.show()
```
