### demonstrate dimensionality_reduction function

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
df_reduced = functions.dimensionality_reduction(numeric_df, 2, [])

# For visualization:
fig, ax = plt.subplots()
ax.scatter(df_reduced['PC1'], df_reduced['PC2'], alpha=0.5)
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')

plt.show()
```

## Visualize the Reduced Data

### Compare Cities

```python
agg_data = functions.group_and_aggregate_data(df.drop(columns='ballot_code'), 'city_name', 'sum')

sparse_remove_agg = functions.remove_sparse_columns(agg_data, 1000)

df_reduced_agg = functions.dimensionality_reduction(sparse_remove_agg, 2, [])

# For visualization:
fig, ax = plt.subplots()
ax.scatter(df_reduced_agg['PC1'], df_reduced_agg['PC2'], alpha=0.5)
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()

print('We estimate that the number of clusters Visually is : 1')
```

### Compare Parties

```python
data_t = agg_data.T

df_sparse_t = functions.remove_sparse_columns(data_t, 1000)

df_reduced_t = functions.dimensionality_reduction(df_sparse_t, 2, [])

# For visualization:
fig, ax = plt.subplots()
ax.scatter(df_reduced_t['PC1'], df_reduced_t['PC2'], alpha=0.5)
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()

print('We estimate that the number of clusters Visually is : 3')
```
